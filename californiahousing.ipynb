```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
housing_data['ocean_proximity_encoded'] = label_encoder.fit_transform(housing_data['ocean_proximity'])
```

```python
X=housing_data[['latitude','longitude','total_rooms', 'median_income', 'total_bedrooms', 'ocean_proximity_encoded', 'housing_median_age','population']]
y=housing_data['median_house_value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)


print("Mean Squared Error:", mse)
print("Model Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
import numpy as np

def predict_median_house_value(latitude, longitude, total_rooms, median_income, total_bedrooms, ocean_proximity, housing_median_age, population):
    
    ocean_proximity_encoded = label_encoder.transform([ocean_proximity])[0]
    
    
    input_features = np.array([[latitude, longitude, total_rooms, median_income, total_bedrooms, ocean_proximity_encoded, housing_median_age, population]])
    
   
    predicted_value = model.predict(input_features)
    
    return predicted_value[0]


predicted_value = predict_median_house_value(
    latitude=37.88, 
    longitude=-122.23, 
    total_rooms=1500, 
    median_income=4.5, 
    total_bedrooms=300, 
    ocean_proximity='NEAR BAY', 
    housing_median_age=20, 
    population=500
)

print("Predicted Median House Value:", predicted_value)

```
